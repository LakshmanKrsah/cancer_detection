```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Correct file path handling
file_path = r'C:\Users\Lakshman kumar sah\Desktop\HEALTH\notebook\data\GroundTruth.csv'

# Load the dataset
df = pd.read_csv(file_path)

```

```python
# Display dataset overview
print("Dataset preview:")
display(df.head())
```

```python
# Dataset information
print("\nDataset information:")
df.info()
```

```python
# Summary statistics
print("\nSummary statistics:")
display(df.describe())
```

```python
# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Diagnoses columns
diagnosis_columns = df.columns[1:]
```

```python
# Distribution of each diagnosis
print("\nDistribution of diagnoses (counts):")
diagnosis_counts = df[diagnosis_columns].sum()
display(diagnosis_counts)
```

```python
# Plot distribution of diagnoses
plt.figure(figsize=(10, 6))
sns.barplot(x=diagnosis_counts.index, y=diagnosis_counts.values, palette="viridis")
plt.title("Distribution of Diagnoses")
plt.xlabel("Diagnosis")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Check correlation between diagnoses
print("\nCorrelation Heatmap:")
correlation_matrix = df[diagnosis_columns].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Diagnoses")
plt.show()
```

```python
# Analyze the distribution of specific diagnoses (e.g., MEL)
print("\nAnalyzing MEL Distribution:")
sns.histplot(df['MEL'], bins=10, kde=True, color='blue')
plt.title("MEL Distribution")
plt.xlabel("MEL")
plt.ylabel("Frequency")
plt.show()
```

```python
# Analyze image-wise diagnosis occurrence
df['diagnosis_count'] = df[diagnosis_columns].sum(axis=1)

print("\nImage-wise diagnosis counts:")
sns.countplot(x='diagnosis_count', data=df, palette="Set2")
plt.title("Number of Diagnoses Per Image")
plt.xlabel("Number of Diagnoses")
plt.ylabel("Count")
plt.show()
```

```python
# Save cleaned dataset if needed (e.g., remove unnecessary columns or preprocess further)
cleaned_file_path = "Cleaned_GroundTruth.csv"
df.to_csv(cleaned_file_path, index=False)
print(f"\nCleaned dataset saved as '{cleaned_file_path}'")
```
